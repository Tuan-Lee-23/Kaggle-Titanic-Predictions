### SibSp and Parch

```python
# Creat a new feature called FamilySize
for dataset in full_data:
    dataset['FamilySize'] = dataset['SibSp'] + dataset['Parch'] + 1
train.head()
```

```python
train[["FamilySize", 'Survived']].groupby("FamilySize").mean()
```

```python
# Create a new feature called IsAlone (Binary Feature)
for dataset in full_data:
    dataset['IsAlone'] = 0
    dataset.loc[dataset['FamilySize'] == 1, "IsAlone"] = 1
train.head()
```

```python
train[['IsAlone', 'Survived']].groupby("IsAlone").mean()
```

### Embarked

```python
# Fill missing value with the mode
for dataset in full_data:
    dataset['Embarked'].fillna("S", inplace = True)
train[['Embarked', 'Survived']].groupby('Embarked').mean()
```

```python
train.isnull().any()
```

### Fare

```python
#Checking missing values in "Fare" column
train["Fare"].isnull().any()
test["Fare"].fillna(test["Fare"].median(), inplace = True)
```

```python
print(train.isnull().any())
print("-----------------")
print(test.isnull().any())
```

### Age

```python
# Checking missing values
train["Age"].isnull().any()
```

```python
train.isnull().sum()
```

```python
# Fill missing values in Age
for dataset in full_data:
    age_mean = dataset["Age"].mean()
    age_std = dataset["Age"].std()
    age_null_count = dataset["Age"].isnull().sum()
    
    #Then we randomly generate number for missing values
    age_null_random_list = np.random.randint(age_mean - age_std, age_mean +  age_std, size = age_null_count)
    
    #Then we assign these number
    dataset['Age'][np.isnan(dataset['Age'])] = age_null_random_list
    
    # Ep kieu thanh integer = Round these numbers
    dataset['Age'] = dataset['Age'].astype(int)
   
```

### Age

```python
import re
def getTitle(name):
    title_search = re.search('([A-Za-z]+)\.', name)
    
    #If the title exist, extract and return it
    if title_search:
        return title_search.group(1)
    return ""

for dataset in full_data:
    dataset['Title'] = dataset['Name'].apply(getTitle)
train.head()
```

```python
pd.crosstab(train["Title"], train['Sex'])
```

```python
# Processing The Title
for dataset in full_data:
    
    # Put Rare Title into "Rare" label
    dataset["Title"] = dataset["Title"].replace(['Dona', 'Don','Capt', 'Col', 'Dr', 'Countess', 'Jonkheer', 'Lady', 'Major', 'Rev', 'Sir'], 'Rare')
    
    # Mlle is Miss in French
    dataset["Title"] = dataset["Title"].replace("Mlle", "Miss")
    
    # Mme is Mr in French
    dataset["Title"] = dataset["Title"].replace("Mme", "Mrs")
    
    #Ms is Miss 
    dataset['Title'] = dataset['Title'].replace("Ms", "Miss")
    
train['Title']
```

```python
print(train[['Title', 'Survived']].groupby("Title").count())
sns.countplot(x = "Survived", hue = "Title", data = train)
```
